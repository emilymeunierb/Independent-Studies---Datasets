```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


from google.colab import files
uploaded = files.upload()

winter_df = pd.read_csv("Updated_winter_warmers_Recipes (1).csv")


#Categorize recipes by ingredient count
def categorize_complexity(count):
    if count <= 10:
        return "Simple (≤10)"
    elif count <= 20:
        return "Moderate (11–20)"
    else:
        return "Complex (>20)"

winter_df['complexity'] = winter_df['ingredient_count'].apply(categorize_complexity)

#Count recipes in each category
complexity_counts = winter_df['complexity'].value_counts().reindex(["Simple (≤10)", "Moderate (11–20)", "Complex (>20)"])

#Create plot yellow bar chart
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.barplot(x=complexity_counts.index, y=complexity_counts.values, color="gold")
plt.title("Recipe Complexity Based on Ingredient Count")
plt.xlabel("Recipe Type")
plt.ylabel("Number of Recipes")
plt.tight_layout()
plt.show()

```

```python
#Define the ingredient columns
ingredients = [
    "contains_egg", "contains_milk", "contains_butter",
    "contains_cheese", "contains_flour", "contains_sugar"
]

#Count how many times each ingredient appears
ingredient_totals = winter_df[ingredients].sum().sort_values(ascending=False)

# Create plot yellow bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=ingredient_totals.values, y=ingredient_totals.index, color="purple")
plt.title("Frequency of Ingredient Usage in Winter Recipes")
plt.xlabel("Number of Recipes")
plt.ylabel("Ingredient")
plt.tight_layout()
plt.show()
```
