```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()

sushi_df = pd.read_csv("Updated_sushi_recipes.csv")

#Create plot histogram of ingredient count
plt.figure(figsize=(8, 5))
sns.histplot(sushi_df["ingredient_count"], bins=10, kde=False, color="lightseagreen")
plt.title("Distribution of Ingredient Count in Sushi Recipes")
plt.xlabel("Number of Ingredients")
plt.ylabel("Number of Recipes")
plt.tight_layout()
plt.show()
```

```python
#Define ingredient columns
ingredients = [
    "contains_egg", "contains_milk", "contains_butter",
    "contains_cheese", "contains_flour", "contains_sugar"
]

#Count how many times each appears
ingredient_totals = sushi_df[ingredients].sum().sort_values(ascending=False)

#Create plot bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=ingredient_totals.values, y=ingredient_totals.index, color="brown")
plt.title("Frequency of Ingredient Usage in Sushi Recipes")
plt.xlabel("Number of Recipes")
plt.ylabel("Ingredient")
plt.tight_layout()
plt.show()
```
