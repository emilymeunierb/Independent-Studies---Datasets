```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()

veg_df = pd.read_csv("Updated_vegetarian_recipes_.csv")

#Create scatter Plot - Ingredient Count vs. Uniqueness
plt.figure(figsize=(8, 6))
sns.scatterplot(data=veg_df, x="ingredient_count", y="uniqueness", color="red")
plt.title("Ingredient Count vs. Uniqueness in Vegetarian Recipes")
plt.xlabel("Number of Ingredients")
plt.ylabel("Uniqueness Score")
plt.tight_layout()
plt.show()
```

```python
#Define ingredient columns
ingredients = [
    "contains_egg", "contains_milk", "contains_butter",
    "contains_cheese", "contains_flour", "contains_sugar"
]

#Count how often each ingredient appears
ingredient_totals = veg_df[ingredients].sum().sort_values(ascending=False)

#Create plot bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=ingredient_totals.values, y=ingredient_totals.index, color="mediumseagreen")
plt.title("Frequency of Ingredient Usage in Vegetarian Recipes")
plt.xlabel("Number of Recipes")
plt.ylabel("Ingredient")
plt.tight_layout()
plt.show()
```
