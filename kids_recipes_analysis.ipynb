```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()  # Upload file


kids_df = pd.read_csv("Updated_ kids_recipes.csv")
sns.set(style="whitegrid")

#Create a bar chart
plt.figure(figsize=(10, 6))
sns.countplot(data=kids_df, y="chapter", order=kids_df['chapter'].value_counts().index)
plt.title("Number of Recipes by Chapter")
plt.xlabel("Number of Recipes")
plt.ylabel("Recipe Chapter")
plt.tight_layout()
plt.show()
```

```python
ingredient_cols = [
    "contains_egg", "contains_milk", "contains_butter",
    "contains_cheese", "contains_flour", "contains_sugar"
]
ingredients_by_chapter = kids_df.groupby("chapter")[ingredient_cols].sum()

#Plot stacked bar chart

ingredients_by_chapter.plot(kind="bar", stacked=True, figsize=(12, 7))
plt.title("Ingredients Used")
plt.xlabel("Recipe Chapter")
plt.ylabel("Number of Recipes Containing Ingredient")
plt.legend(title="Ingredient", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
