```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()

lunch_df = pd.read_csv("Updated_healthy_schoollunch_recipes.csv")

#Count recipes with and without nuts
nut_counts = lunch_df['contains_nuts'].value_counts()
nut_labels = ['Without Nuts', 'With Nuts']

# Create plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(nut_counts, labels=nut_labels, autopct='%1.1f%%', startangle=140)
plt.title("Proportion of Recipes With vs. Without Nuts")
plt.tight_layout()
plt.show()
```

```python
ingredients = [
    "contains_egg", "contains_milk", "contains_butter",
    "contains_cheese", "contains_flour", "contains_sugar", "contains_nuts"
]

#Count how many times each ingredient appears
ingredient_totals = lunch_df[ingredients].sum().sort_values(ascending=False)

#Create plot bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=ingredient_totals.values, y=ingredient_totals.index)
plt.title("Frequency of Ingredient Usage in Lunch Recipes")
plt.xlabel("Number of Recipes")
plt.ylabel("Ingredient")
plt.tight_layout()
plt.show()
```
